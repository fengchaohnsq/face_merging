```python
import cv2
import numpy as np
import math
from PIL import Image
```

```python
# 从text文本中读取特征点
def read_points(path) :
    # Create an array of points.
    points = []
    # Read points
    with open(path) as file :
        for line in file :
            x, y = line.split()
            points.append((int(x), int(y)))
    return points

#根据特征点正置图片
def face_rotation(file_path,point_path,save_path,flag):
    flag=1
    points = read_points(point_path)
    a=points[27][0]-points[30][0]#正负需要留意会影响旋转角度
    b=points[30][1]-points[27][1]
    alpha= math.atan(a/b)
    angle=alpha/math.pi*180
    #读取图像
    im = Image.open(file_path)
    # 指定逆时针旋转的角度
    print(angle)
    im_rotate = im.rotate(angle*flag) 
    im_rotate.save(save_path)
    
#根据特征点中心位置扣图保存
def shape_face(file_path,point_path,save_path):
    img=cv2.imread(save_path)
    hsv=cv2.cvtColor(img,cv2.COLOR_BGR2HSV)
    black=np.array([0,0,0])
    mask = cv2.inRange(hsv,black,black,255)
    
    #腐蚀膨胀
    erode=cv2.erode(mask,None,iterations=1)
    dilate=cv2.dilate(erode,None,iterations=1)
    
    #将RGB图转化成二值图
    imgray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    ret,thresh = cv2.threshold(imgray,10,255,cv2.THRESH_BINARY)
    contours = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_NONE)
    cnt = contours[0]
    
    # #boundingRect框图法
    x, y, w, h = cv2.boundingRect(cnt)
#     cv2.rectangle(img, (x, y), (x+w+10, y+h+10), (0, 0, 255), 3)
    img=img[y:y+h,x:x+w]
                     
    cv2.namedWindow("res",0)
    cv2.imshow('res',img)
    cv2.imwrite(save_path,img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
filename='img0.7.jpg'
file_path='/home/fengchao/桌面/merge results/'+filename
point_path='/home/fengchao/桌面/MonkeyFaceMerging/merge points/morph_points.txt'
save_path='/home/fengchao/桌面/merge results/rotation/'+filename
flag=1#控制角度正负
face_rotation(file_path,point_path,save_path,flag)
shape_face(file_path,point_path,save_path)

# 遍历替换
# for i in range(rows):
#     for j in range(cols):
#         if dilate[i,j]==0:#0代表黑色的点
#             img_new[i-500,j-200]=img[i,j]#此处替换颜色，为BGR通道


# face_rows=0
# rows_len=[]
# cols_len=[]
# t=0
# numr=0
# numc=0
# #计算行数
# for i in range(rows):
#     for j in range(cols):
#         if dilate[i,j]==0:
#             if len(rows_len)==0:
#                 rows_len.append(i)
#             if rows_len[numr]!=i:
#                 rows_len.append(i)
#                 numr=numr+1
# flag=1
# #计算最大列数
# for j in range(cols):
#     for i in range(rows):
#         if dilate[i,j]==0:
#             if len(cols_len)==0:
#                 cols_len.append(j)
#             for k in range(len(cols_len)):
#                 if(cols_len[k]==j):
#                     continue
#                     flag=0
#             if flag==1:
#                 cols_len.append(j)
                
                
# print(rows)                
# print(numr)#行数只计算行 行数固定列数不固定
# print(numc)#列数计算的是所有的列
            

#直接截图法
# points = read_points(point_path)
# img=img[points[29][0]-1000:points[29][0]+1000,points[29][1]-1000:points[29][1]+1000]
# cv2.namedWindo w("res",0)
# cv2.imwrite(save_path,img)
# cv2.imshow('res',img)
# cv2.waitKey(0)
# cv2.destroyAllWindows()
```
